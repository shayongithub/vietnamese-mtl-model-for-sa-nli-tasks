```python
df = pd.read_csv(r"C:\Users\Dell\Downloads\ggnews.csv", encoding ="utf-8-sig")
df.head()
```

```python
df = df.drop('author', axis=1)
df.head()
```

```python
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
# nltk.download('punkt')
# Character counts
df["char_count_in_content"] = df["content"].map(lambda x: len(x))
# Word counts
df["word_count_in_content"] = df["content"].map(lambda x: len(x.split()))
# Sentence counts
df["sent_count"] = df["content"].map(lambda x: len(sent_tokenize(x)))
# String counts
# df["hashtag_count"] = df["content"].map(lambda x: x.count("#"))
# Average word length
df["avg_word_len"] = df["content"].map(lambda x: np.mean([len(w) for w in str(x).split()]))
# Average sentence length
df["avg_sent_len"] = df["content"].map(lambda x: np.mean([len(w.split()) for w in sent_tokenize(x)]))
```

```python
import pandas as pd
from underthesea import word_tokenize

# Tải tập stopwords tiếng Việt từ file txt
with open('vietnamese_stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = set([w.strip() for w in f.readlines()])

def preprocess_text(text):
    # chuyển đổi các từ thành lowercase
    text = text.lower()
    # loại bỏ các dấu câu
    text = "".join(c for c in text if c not in ('!', '.', ',', '?', ':', ';', '-', '(', ')', '[', ']', '{', '}'))
    # tokenize câu
    tokens = word_tokenize(text)
    # loại bỏ các stopwords
    tokens = [token for token in tokens if token not in stopwords]
    return tokens



# Áp dụng hàm preprocess_text vào cột 'content' trong dataframe và tạo cột mới 'text_cleaned'
df['text_cleaned'] = df['content'].apply(preprocess_text)

print(df)
```

```python
from collections import Counter
def find_common(corpus,n=10):
    word_count = Counter(corpus)
    top_words = word_count.most_common(n)
    return top_words
df["most_common"]=df['text_cleaned'].apply(find_common)
```
